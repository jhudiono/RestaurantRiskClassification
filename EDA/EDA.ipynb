```python
import pandas as pd
```

```python
def txt_to_dataframe(filename, columns=["venue", "count"]):
    names = ""
    with open(filename) as file:
        names = file.read().strip()
    names = names.split("\n")[2:-1]
    count = {}
    print(len(names))
    for i, n in enumerate(names):
        n = n.split("|")
        try:
            count[n[0].strip()] = n[1].strip()
        except:
            print(i, n)
    return count, pd.DataFrame(list(count.items()), columns=columns)
```
